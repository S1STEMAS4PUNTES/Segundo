```python
import matplotlib.pyplot as plt     #Biblioteca para la generación de gráficos    
import seaborn as sns; sns.set()    #Biblioteca para la generación de gráficos  estadisticos
import numpy as np                  #Biblioteca que da soporte para crear vectores y matrices
```

```python
x = np.array([21.04, 16.00, 24.00, 14.16, 30.00, 19.85, 15.34, 14.27, 13.80, 14.94])
y = np.array([39.9900, 32.9900, 36.9000, 23.2000, 53.9900, 29.9900, 31.4900, 19.8999, 21.2000, 24.2500])

plt.scatter(x, y)
plt.xlabel("Tamaño $ft^2$ (x100)", size = 16,)
plt.ylabel("Precio en dolares (10000)", size = 16)
plt.title("Predicción precio de casa", 
          fontdict={'family': 'serif', 
                    'color' : 'darkblue',
                    'weight': 'bold',
                    'size': 18})

plt.grid(True)
```

```python
# Resolviendo de forma analítica

n = len(x)
X = np.zeros((n,2)) 
Theta = np.zeros((2,1))

X[:,0] = 1
X[:,1] = x
# Encontrar los valores b y m

Theta = np.dot(np.dot(np.linalg.pinv(np.dot(np.transpose(X),X)), np.transpose(X)),y)
print(Theta)
```

```python
# Graficar 

y1 = Theta[1]*x + Theta[0]
plt.plot(x, y1, 'g', label = "Regresión lineal analítica")
plt.scatter(x,y, label = "Datos")
plt.legend()
plt.xlabel("Tamaño $ft^2$ (X100)", size = 16,)
plt.ylabel("Precio en dolares (x10000)", size = 16)
plt.title("Predicción precio de casa", 
          fontdict={'family': 'serif', 
                    'color' : 'darkblue',
                    'weight': 'bold',
                    'size': 18})
plt.show()
```

```python
T_pred = 28 #Tamaño de casa para predecir su valor
y_pred = Theta[0] + Theta[1]*T_pred
print("Para un tamaño de:", T_pred, "ft^2 (X100)", "el precio es de:", y_pred, "dolares (x10000)")
```
